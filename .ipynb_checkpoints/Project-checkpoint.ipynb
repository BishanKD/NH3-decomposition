# Machine Learning Modelling of Catalytic Ammonia Decomposition Process

### 1) Importing libraries and dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./nh3_decomposition.csv')
df.head()
```

### 2) Preprocessing and Exploratory Data Analysis (EDA)

```python
df.columns
```

```python
df = df.drop('DOI', axis=1)
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(3, 3, figsize=(17, 12))
ax = ax.flatten()

for i, col in enumerate(df.columns):
    ax[i].hist(df[col], bins=20, edgecolor='black')
    ax[i].set_title("Distribution of "+ col)

plt.subplots_adjust(hspace=0.4)
fig.suptitle('Distribution of all variables', fontsize=20)

plt.show()
```

```python
sns.heatmap(df.corr(method='spearman'), annot=True, center=0, cmap='YlGnBu', fmt=".2f", linewidths=0.5)

plt.title('Spearman Correlation Matrix');
```

```python
fig, ax = plt.subplots(2, 4, figsize=(13, 6))
ax = ax.flatten()

for i, col in enumerate(df.columns[:-1]):
    ax[i].scatter(x=df[col], y=df['Ammonia conversion (%)'])
    ax[i].set_xlabel(col)

plt.subplots_adjust(hspace=0.5, wspace=0.4)
fig.suptitle('Visualization of different input features against Ammonia production', fontsize=16)

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('Ammonia conversion (%)', axis=1)
y = df['Ammonia conversion (%)']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### 3) Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=range(1, len(pca.explained_variance_ratio_)+1),
             y=pca.explained_variance_ratio_, marker='o')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=range(1, len(pca.explained_variance_ratio_.cumsum())+1),
             y=pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title('Explained Variance by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative explained Variance Ratio')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.ravel()

fig.suptitle('Scatter Plots of Principal Components vs Ammonia Conversion', fontsize=13, y=1)
for i in range(4):
    axes[i].scatter(X_pca[:, i], y)
    axes[i].set_xlabel(f'Principal Component {i+1}')
    axes[i].set_ylabel('Ammonia conversion (%)')
    axes[i].set_title(f'PC{i+1} vs Ammonia conversion')

plt.tight_layout()
plt.show()
```

### 4) Model Fitting

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Regressor:")
print("R2 Score:", r2_score(y_test, y_pred_rf))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
print("MAE:", mean_absolute_error(y_test, y_pred_rf))
```

```python
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

print("\nDecision Tree Regressor:")
print("R2 Score:", r2_score(y_test, y_pred_dt))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_dt)))
print("MAE:", mean_absolute_error(y_test, y_pred_dt))
```

```python
svm = SVR()
svm.fit(X_train, y_train)
y_pred_svm = svm.predict(X_test)

print("\nSupport Vector Regressor:")
print("R2 Score:", r2_score(y_test, y_pred_svm))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_svm)))
print("MAE:", mean_absolute_error(y_test, y_pred_svm))
```

#### Plotting performance metrics

```python
models = ['Random Forest', 'Decision Tree', 'SVM']
r2 = [r2_score(y_test, y_pred_rf), 
      r2_score(y_test, y_pred_dt), 
      r2_score(y_test, y_pred_svm)
     ]
rmse = np.sqrt([
    mean_squared_error(y_test, y_pred_rf),
    mean_squared_error(y_test, y_pred_dt),
    mean_squared_error(y_test, y_pred_svm)
])
mae = [
    mean_absolute_error(y_test, y_pred_rf),
    mean_absolute_error(y_test, y_pred_dt),
    mean_absolute_error(y_test, y_pred_svm)
]

plt.figure(figsize=(8, 5))
plt.plot(models, r2, marker='o', label='R² Score')
plt.plot(models, rmse, marker='o', label='RMSE')
plt.plot(models, mae, marker='o', label='MAE')

for i in range(len(models)):
    plt.text(models[i], r2[i] - 1.5, f"{r2[i]:.2f}", ha='center')
    plt.text(models[i], rmse[i] - 1.5, f"{rmse[i]:.2f}", ha='center')
    plt.text(models[i], mae[i] - 1.5, f"{mae[i]:.2f}", ha='center')

plt.title('Model Evaluation Metrics')
plt.ylabel('Metric Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 5) Hyperparameter Tuning with 5-fold Cross Validation

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}
param_grid_dt = {
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
param_grid_svr = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['rbf', 'linear']
}

gs_rf  = GridSearchCV(RandomForestRegressor(random_state=42),
                      param_grid_rf, cv=5,
                      scoring='neg_mean_squared_error', n_jobs=-1)
gs_dt = GridSearchCV(DecisionTreeRegressor(random_state=42),
                     param_grid_dt, cv=5,
                     scoring='neg_mean_squared_error', n_jobs=-1)
gs_svr = GridSearchCV(SVR(),
                      param_grid_svr, cv=5,
                      scoring='neg_mean_squared_error', n_jobs=-1)


gs_rf.fit(X_train, y_train)
gs_dt.fit(X_train, y_train)
gs_svr.fit(X_train, y_train)

for name, gs in [('Random Forest', gs_rf), ('Decision Tree', gs_dt), ('SVR', gs_svr)]:
    best_rmse = np.sqrt(-gs.best_score_)
    print(f"{name} best params: {gs.best_params_}")
    print(f"{name} CV RMSE: {best_rmse:.3f}\n")
```

#### Plotting RMSE before and after tuning

```python
rmse_after = np.sqrt([
    -gs_rf.best_score_,
    -gs_dt.best_score_,
    -gs_svr.best_score_
])

models = ['Random Forest', 'Decision Tree', 'SVM']
x = np.arange(len(models))

plt.figure(figsize=(10, 6))

plt.plot(x, rmse, marker='s', label='RMSE Before')

plt.plot(x, rmse_after, marker='s', linestyle='--', label='RMSE After')

for i in range(len(models)):
    plt.text(x[i] - 0.15, rmse[i], f"{rmse[i]:.3f}", color='orange', fontsize=10)
    plt.text(x[i] + 0.05, rmse_after[i], f"{rmse_after[i]:.3f}", color='orange', fontsize=10)

plt.xticks(x, models)
plt.title('RMSE Before and After Tuning')
plt.ylabel('RMSE Value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 6) Exporting the tuned models

```python
import pickle

with open('./models/rf_tuned.pkl', 'wb') as f:
    pickle.dump(gs_rf.best_estimator_, f)

with open('./models/dt_tuned.pkl', 'wb') as f:
    pickle.dump(gs_dt.best_estimator_, f)

with open('./models/svr_tuned.pkl', 'wb') as f:
    pickle.dump(gs_svr.best_estimator_, f)
```
